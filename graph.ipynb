```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data for WUE by Treatment
data = {
    'Treatment': ['T1', 'T2', 'T3', 'T5', 'T6'],
    'WUE (kg/ha/mm)': [13.26, 13.26, 41.02, 23.67, 42.13]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style
sns.set(style="whitegrid")

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Treatment', y='WUE (kg/ha/mm)', data=df, palette='viridis')
plt.title('Water Use Efficiency (WUE) by Treatment')
plt.ylabel('WUE (kg/ha/mm)')
plt.xlabel('Treatment')
plt.ylim(0, 50)  # Adjust the y-axis limits for better visualization
plt.savefig("water_use_efficiency.png", dpi=400)
# Show the plot
plt.show()
```
